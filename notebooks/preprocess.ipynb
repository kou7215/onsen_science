## データ非欠損率

データがどれだけ欠損してないか.

座標データは75%くらいは生きてそう.

座標データが残っているサンプルのみを抽出.

```python
def NA_counter(x_df):
    n_nan = x_df.isnull().sum()
    x_df_dtype = x_df.dtypes
    percent_nan = n_nan/len(x_df)
    return pd.DataFrame([n_nan, percent_nan,x_df_dtype]).rename({0:'n_nan', 1:'%', 2:'dtype'}) 

na_df = NA_counter(df)
na_df
```
